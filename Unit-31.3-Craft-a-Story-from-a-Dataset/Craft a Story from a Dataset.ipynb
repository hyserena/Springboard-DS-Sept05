## Storytelling 
I want to compart the environmental impact of plant-based vs. animal-based products and examine the potential benefits of dietary shifts at a population level.

Hypothesis: Transitioning to plant-based diets can substantially mitigate the environmental impacts associated with food production.

```python
plant_based = ['Wheat & Rye (Bread)', 'Maize (Meal)', 'Barley (Beer)', 'Oatmeal', 'Rice', 'Potatoes', 'Cassava', 'Cane Sugar', 'Beet Sugar', 'Other Pulses', 'Peas', 'Nuts', 'Groundnuts', 'Soymilk', 'Tofu', 'Soybean Oil', 'Palm Oil', 'Sunflower Oil', 'Rapeseed Oil', 'Olive Oil']
animal_based = ['Beef (beef herd)', 'Beef (dairy herd)', 'Lamb & Mutton', 'Pork', 'Poultry', 'Milk', 'Cheese', 'Eggs', 'Fish (farmed)']

food_production['Category'] = food_production['Food product'].apply(lambda x: 'Animal-Based' if x in animal_based else 'Plant-Based')

average_impacts = food_production.groupby('Category').mean()

metrics_to_compare = ['Total_emissions', 'Freshwater withdrawals per kilogram (liters per kilogram)', 'Land use per kilogram (m² per kilogram)']

fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)
for i, metric in enumerate(metrics_to_compare):
    axes[i].bar(average_impacts.index, average_impacts[metric], color=['skyblue', 'salmon'])
    axes[i].set_title(metric.replace('_', ' '))
    axes[i].set_xlabel('Category')
    if i == 0:
        axes[i].set_ylabel('Average Value')

plt.tight_layout()
plt.show()

average_impacts[metrics_to_compare]
```

#### Comparative Analysis: Plant-Based vs. Animal-Based Products
The analysis comparing the environmental impact of plant-based and animal-based products across key metrics reveals significant differences:

- Total Emissions: Animal-based products have an average total emission of about 19.83 CO₂eq per kilogram, which is significantly higher than the average for plant-based products, at approximately 3.28 CO₂eq per kilogram. This stark difference highlights the much higher carbon footprint associated with the production of animal-based foods.
- Freshwater Withdrawals: The average freshwater withdrawals for animal-based products are estimated at 2352.96 liters per kilogram, far exceeding the 611.88 liters per kilogram for plant-based products. This indicates that animal-based products typically require much more water, contributing to greater environmental strain in water-scarce regions.
- Land Use: Similarly, land use per kilogram for animal-based products averages 121.53 m², compared to just 8.43 m² for plant-based products. This vast difference underscores the inefficiency of animal-based food production in terms of land use, impacting deforestation, biodiversity loss, and more.

#### Insights and Potential Benefits of Dietary Shifts
- Environmental Impact Reduction: Shifting dietary preferences from animal-based to plant-based foods could significantly reduce the overall environmental impact, notably in greenhouse gas emissions, freshwater use, and land use.
- Sustainability: A population-level shift towards plant-based diets could contribute to more sustainable food systems, alleviating pressure on natural resources and helping to combat climate change.
- Health and Economic Benefits: Besides environmental advantages, transitioning to plant-based diets is often associated with health benefits and could potentially lead to reduced healthcare costs and economic savings.

```python
average_impact_metrics = food_production.groupby('Category').mean()[['Total_emissions', 'Freshwater withdrawals per kilogram (liters per kilogram)', 'Land use per kilogram (m² per kilogram)']]

avg_emissions_animal = average_impact_metrics.loc['Animal-Based', 'Total_emissions']
avg_emissions_plant = average_impact_metrics.loc['Plant-Based', 'Total_emissions']

emissions_reduction_per_kg = avg_emissions_animal - avg_emissions_plant

daily_emissions_reduction = emissions_reduction_per_kg * 0.5

average_impact_metrics, emissions_reduction_per_kg, daily_emissions_reduction
```
